```python
import pandas as pd
```

```python
# Bring & Read Data
df = pd.read_excel(r"covid-19.xlsx")
df
```

```python
# Exploration Data Analysis
df.info()
```

```python
df.sample(10)
```

```python
df.describe()
```

```python
# Print the number of empty or duplicate values
df.isnull().sum()
```

```python
print(f"Number of duplicate = {df.duplicated().sum()}")
```

```python
# Extract KPIs

Total_Cases = df['Cases'].sum()
print(f"Total Cases = {Total_Cases}")

Total_Deaths = df['Deaths'].sum()
print(f"Total Desths = {Total_Deaths}")

Total_Active = df['Active'].sum()
print(f"Total Active = {Total_Active}")

Total_Recovered = df['Recovered'].sum()
print(f"Total Recovered = {Total_Recovered}")

Deaths_Rate = Total_Deaths / Total_Cases
print(f"Deaths Rate = {Deaths_Rate}")

Recovery_Rate = Total_Recovered / Total_Cases
print(f"Recovery Rate = {Recovery_Rate}")
```

```python
# Top 10 Counties Cases and Deaths as well as Continent
top10_Deaths_Countries = df.groupby('Country_Region')['Deaths'].sum().sort_values(ascending=False).head(10)
print(f"Top 10 Countries where Deaths : {top10_Deaths_Countries}")
print("-------------------")
top10_Cases_Countries = df.groupby('Country_Region')['Cases'].sum().sort_values(ascending=False).head(10)
print(f"Top 10 Countries where Cases : {top10_Cases_Countries}")
```

```python
Cases_Continent = df.groupby('WHO Region')['Cases'].sum().sort_values(ascending=False)
print(f"Continent where Cases : {Cases_Continent}")
print("-------------------")
Deaths_Continent = df.groupby('WHO Region')['Deaths'].sum().sort_values(ascending=False)
print(f"Continent where Deaths : {Deaths_Continent}")
```

```python
# The rate of Cases, Deaths & Active over time
Cases_Over_Time = df.groupby('Date')['Cases'].sum()
print(f"Cases over time : {Cases_Over_Time}")
```

```python
Deaths_Over_Time = df.groupby('Date')['Deaths'].sum()
print(f"Deaths over time : {Deaths_Over_Time}")
```

```python
Active_Over_Time = df.groupby('Date')['Active'].sum()
print(f"Active over time : {Active_Over_Time}")
```
